4. Проверить оставшиеся условия применимости линейной регрессии

```python
y_pred = b1 * X1 + b0
resid = y_pred - Y1
from scipy import stats
stats.shapiro(resid) #нет отличий от нормального распределения - H0, p-value > 5%, потому принимаем H0
```

```python
df1 = 2 - 1
df2 = len(X1) - 2
MSF = np.sum((y_pred - np.mean(Y1)) ** 2 / df1)
MSO = np.sum(resid ** 2 / df2)
F = MSF / MSO
F
```
